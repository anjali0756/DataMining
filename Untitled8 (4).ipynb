```python
# Import necessary libraries
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# Step 1: Load the dataset
# Replace 'dataset.csv' with the name of your CSV file
file_name = 'dataset.csv'  # Ensure the file is uploaded to the runtime
dataset = pd.read_csv(file_name, sep=';')
```

```python
# Step 2: Inspect the data
print("\nFirst few rows of the dataset:")
print(dataset.head())
print("\nDataset Information:")
print(dataset.info())
print("\nStatistical Summary:")
print(dataset.describe())
```

```python
# Step 3: Visualizations
sns.set(style="whitegrid")  # Set the style for Seaborn
```

```python
# 1. Line Plot: Average quality vs. alcohol content
plt.figure(figsize=(10, 6))
sns.lineplot(data=dataset, x="alcohol", y="quality", marker='o', ci=None)
plt.title("Average Quality vs. Alcohol Content")
plt.xlabel("Alcohol Content (%)")
plt.ylabel("Average Quality")
plt.grid(True)
plt.show()
```

```python
# 2. Scatter Plot: Alcohol vs. Density, colored by quality
plt.figure(figsize=(10, 6))
sns.scatterplot(data=dataset, x="alcohol", y="density", hue="quality", palette="viridis")
plt.title("Alcohol vs. Density")
plt.xlabel("Alcohol Content (%)")
plt.ylabel("Density")
plt.show()
```

```python
# 3. Bar Plot: Average sulphates by quality
plt.figure(figsize=(10, 6))
sns.barplot(data=dataset, x="quality", y="sulphates", palette="mako")
plt.title("Average Sulphates by Quality")
plt.xlabel("Quality")
plt.ylabel("Average Sulphates")
plt.show()
```

```python
# 4. Histogram: Distribution of alcohol content
plt.figure(figsize=(10, 6))
sns.histplot(data=dataset, x="alcohol", bins=20, kde=True, color="coral")
plt.title("Distribution of Alcohol Content")
plt.xlabel("Alcohol Content (%)")
plt.ylabel("Frequency")
plt.show()
```

```python
# 5. Pie Chart: Proportion of wine qualities
quality_counts = dataset["quality"].value_counts()
plt.figure(figsize=(8, 8))
plt.pie(quality_counts, labels=quality_counts.index, autopct='%1.1f%%', startangle=140, colors=sns.color_palette("pastel"))
plt.title("Proportion of Wine Qualities")
plt.show()
```
